```python
from utils.hnsw_module import *
import pandas as pd


data = pd.read_csv('data/embeddings_films.csv', usecols=[i for i in range(2, 386)])
len(data)
train_data = data[:20000]
test_data = data[20000:]
```

```python
import matplotlib.pyplot as plt
# Plot two specific dimensions
dim_x = 0  # Change this to the index of the dimension you want on the x-axis
dim_y = 1  # Change this to the index of the dimension you want on the y-axis

plt.figure(figsize=(12, 6))

# Plot training data
plt.scatter(train_data.iloc[:, dim_x], train_data.iloc[:, dim_y], s=1, label='Training Data')

# Plot test data
plt.scatter(test_data.iloc[:, dim_x], test_data.iloc[:, dim_y], s=1, label='Test Data', alpha=1)

plt.xlabel(f'Dimension {dim_x+1}')
plt.ylabel(f'Dimension {dim_y+1}')
plt.title(f'Plot of Dimension {dim_x+1} vs Dimension {dim_y+1}')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

# Calculate distances from the reference point to all other points in train_data
distances = cdist(data[20000:].values[10].reshape(1,-1), data[:20000].values , metric='euclidean').flatten()

# Plot a histogram of the distances
plt.figure(figsize=(12, 6))
plt.hist(distances, bins=50, color='blue', alpha=0.7)
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.title('Histogram of Distances from Reference Point to All Points in Train Data')
plt.grid(True)
plt.show()
```

```python
dists = []
for i in range(384):
    dists.append(np.linalg.norm(train_data.iloc[15000:20000,i]))
```

```python
plt.figure(figsize=(12, 6))
plt.scatter([i for i in range(384)],dists)
plt.xlabel('Dimensions')
plt.ylabel('Abs value')
plt.title('Abs value of each dimensions ')
plt.show()
```
